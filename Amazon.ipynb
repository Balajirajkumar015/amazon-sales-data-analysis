## Import Libraries & Load Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# display settings
pd.set_option('display.max_columns', None)

# load dataset
df = pd.read_csv("Amazon Sale Report.csv", low_memory=False)


df.head()
```

## Data Cleaning

### Remove unwanted column

```python
df.drop(columns=['Unnamed: 22'], inplace=True)
df.columns
```

### Convert Date column


```python
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y', errors='coerce')
df.head()
```

### Handle Missing Values


```python
df.isnull().sum()
```

```python
df['Amount']=df['Amount'].fillna(0)
```

```python
df.columns =(
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(r'\s+', '_', regex=True)
)
df.columns
```

## Visualizations

### Order Status Count

```python
plt.figure(figsize=(8,5))
sns.countplot(y='status', data=df)
plt.title("Order Status Distribution")
plt.show()
```

### Top 10 States by Sales

```python
state_sales = df.groupby('ship-state')['amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
state_sales.plot(kind='bar')
plt.ylabel("Revenue")
plt.title("Top 10 States by Sales Amount")
plt.show()
```

### Monthly Sales Trend

```python
monthly_sales = df.groupby(df['date'].dt.month)['amount'].sum()

plt.figure(figsize=(8,5))
monthly_sales.plot(marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

### Fulfilment vs Revenue

```python
fulfilment_sales = df.groupby('fulfilment')['amount'].sum()

plt.figure(figsize=(6,4))
fulfilment_sales.plot(kind='bar')
plt.title("Revenue by Fulfilment Type")
plt.show()
```

## Business Insights
